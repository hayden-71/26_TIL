#### 1) 비디오 검색 도구 선언

```python
# 데코레이션 형태로 `에이전트가 사용할 수 있는 도구`로 설정
 # @tool 이라 명시하면 langchain이 함수명, 설명(Docstring), 매개변수 타입 등을 읽어 JSON 설명서를 직접 만들어주고 도구 객체로서 동작하게 설정
 # 또한 OpenAI GPT 모델 뿐만 아니라 구글이나 엔트로픽 등 다른 모델도 @tool로 설정 가능
@tool
def search_videos_tool(query: str) -> str :
    # 에이전트가 사용할 도구에 대한 설명서 (docstring)
    # langchain의 @tool 사용시에는 docstring이 필수적으로 들어가야 함
    # 1) 언제 써야 하는지 명시 / 2) 파라미터 알려주기 / 3) 하지 말아야 할 것 적기
    # (OpenAI에서 tool을 사용할 때 description의 역할)
    """유튜브를 포함한 인터넷 동영상 자료를 검색할 때 사용합니다.
    사용자로부터 검색어(query)를 받아 관련된 동영상 제목과 링크를 반환합니다."""

    try :
        # DDGS()는 덕덕고 서버와 통신하는데 with문이 없다면 원하는 작업이 끝나도 통신이 끊어지지 않고
        # 메모리와 연결포트를 점유하기 때문에 작업이 끝날때 같이 연결도 끊어질 수 있도록 with문으로 작성
        with DDGS() as ddgs :
            results = ddgs.videos(
                keywords=query,
                region='kr-kr',
                timelimit=None,
                max_results=10
            )
            # <DDGS 객체의 다른 메소드>
             # text(): 일반 웹 검색
             # images()
             # videos()
             # news()
             # maps(): 지도/장소 검색

        if not results :
            return f'{query}에 대한 검색 결과가 없습니다..'
        
        output_list = []

        # 반환 형태를 보기 좋게 설정
        for i in results :
            title = i.get('title', '제목 없음')
            link = i.get('content', '링크 없음')
            output_list.append(f'[제목]: {title}\n[링크]: {link}')

        return '\n\n'.join(output_list)
    
    except Exception as e : 
        return f'영상 검색 중 오류 발생: {str(e)}, 오류를 해결하여 다시 시도하세요'
```

```python
search5 = search_videos_tool.invoke('화사 신곡')
print(search5)
```

### 1) Prompt 및 모델 chain 구성

```python
# sqlalchemy: 파이썬 객체를 주면 알아서 접속한 DB의 문법에 맞는 SQL 쿼리로 번역해 전송해주는 모듈 (DB연결시 설치 필요)
!pip install -q sqlalchemy
```

```python
# ChatPromptTemplate: 문자열이 아닌 `메시지 리스트 형태`로 role을 지정한 프롬프트 템플릿
# MessagePlaceholder: 모델의 기억 공간 객체 (대화가 들어갈 빈자리 예약석)
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder

# 과거 대화 메시지를 기록하고 관리하는 클래스 (InMemory.. 저장소에서 꺼내서 -> Placeholder.. 예약석에 넣어주는 자동화 엔진)
 # (다른 메모리처럼 대화 내역을 덧붙이기만 하는 게 아니라, 전체 프롬프트에서 원하는 자리에 과거 대화 내역을 넣을 수 있게 해주고 확장성이 좋음)
from langchain_core.runnables.history import RunnableWithMessageHistory

# 대화 기록들이 머무는 물리적 공간 클래스 (컴퓨터의 메인 메모리 사용하는 대화 기록장. 휘발성!!)
from langchain_core.chat_history import InMemoryChatMessageHistory

# DB를 연동한 채팅 메시지 기록 저장 클래스
from langchain_community.chat_message_histories import SQLChatMessageHistory
```

```python
##### 1. 프롬프트 템플릿 구성
 # 프롬프트는 결국 LLM에게 순차적으로 들어갈 거라, LLM이 받아들이는 대화 시간 순서에 맞추어
 # '시스템 프롬프트' -> '과거 대화 기록' -> '사용자 질의' 순으로 넣어주는 것이 좋음
 # 해당 순서가 에이전트 및 LLM에서 프롬프트 엔지니어링 세계 표준 순서!
prompt = ChatPromptTemplate.from_messages(
    [
        # 시스템 프롬프트 > 모델의 과거 대화 기록 공간 > 사용자 질의
        ('system', '사용자의 질문에 대해 아래 context에 기반하여 답변하시오:\n\n{context}'),
        MessagesPlaceholder(variable_name='chat_history'),  # 대화가 들어갈 '빈자리'! 변수명은 보통 chat_history
        # ('placeholder', '{chat_history}') 와 같은 코드!
        ('human', '{question}')

        # <role 종류>
         # system: 시스템 프롬프트, SystemMessage 객체로 동작
         # human: 사용자 질문, HumanMessage 객체로 동작
         # ai: 모델의 응답, AIMessage 객체로 동작
         # placeholder: 대화 기록 메시지 저장, MessagePlaceholder 객체로 동작
    ]
)

#### 2. LLM 로드
model = ChatOpenAI(model='gpt-4o', api_key=MY_API_KEY)

#### 3. 프롬프트, LLM 체인 구성
my_chain = prompt | model

#### 4. 대화(세션) 기록을 저장할 임시 딕셔너리
store = {}

#### 5. 대화(세션)에 고유 ID 부여
 # 실제 서비스는 많은 유저들이 사용하므로 유저별 다른 값으로 세션을 지정해야 함
 # configurable: langchain에서 지정해둔 예약어로 동적으로 변하는 설정값(세션ID, 유저ID 등)을 담기로 약속된 공간
 # session_id : 마찬가지 예약어로 configurable 내부에서 세션 id를 찾는 key값
config = {'configurable':{'session_id':'user_1'}}

#### 6. 세션 ID(여러 사용자)에 맞추어 메모리를 가져오거나 없으면 새로 만드는 함수
def get_session_history(session_id:str) :
    if session_id not in store :
        store[session_id] = InMemoryChatMessageHistory()  # 물리 메모리 공간 객체 설정
    return store[session_id]

#### 6-1. 세션ID(여러 사용자)에 맞추어 DB 메모리를 연동하는 함수
def get_session_history_db(session_id:str) :
    DB_USER = 'root'    # MariaDB 아이디
    DB_PASS = '1234'    # 비번
    DB_HOST= '127.0.0.1'   # 호스트IP (로컬이면 디폴트)
    DB_PORT= '3306'     # 기본 포트 번호
    DB_NAME='myAgent'   # 미리 생성해둔 DB 이름

    # SQLALchemy용 연결 문자열 (mysql+pymysql 규격 사용)
     # charset=utf8mb4 : utf8 방식 인코딩의 확장 버전
    connection_str=f"mysql+pymysql://{DB_USER}:{DB_PASS}@{DB_HOST}:{DB_PORT}/{DB_NAME}?charset=utf8mb4"

    # DB내
    return SQLChatMessageHistory(
        session_id=session_id,           # 세션id (함수의 매개변수)
        connection=connection_str,       # 연결 문자열
        table_name='chat_history_log'   # 생성될 테이블명
    )

#### 7. 대화 메시지 기록 객체에 체인 융합
chain_with_history = RunnableWithMessageHistory(
    my_chain,                         # 프롬프트 + LLM 체인
    get_session_history_db,               # 세션에 고유 ID를 부여한 메모리 함수
    input_messages_key = 'question',    # 사용자 입력이 들어갈 프롬프트 변수명
    history_messages_key='chat_history'  # 과거 대화 내역이 들어갈 프롬프트 변수명
)
```

```python
question = '라이즈 노래 추천해줘'

answer = chain_with_history.invoke(
    {'question': question,    # 사용자 질문
    'context':search5     # LLM이 참조할 내용 (도구 사용 결과)
    # context에 들어가는 내용은 참조할 텍스트이며, 현재 인터넷 영상 검색 도구
    # (search_videos_tool을 통해 가져온 외부정보이므로 'Web-RAG' 구조로 볼 수 있음)
    },
    config=config     # 세션 id 정보
)

print(answer.content)
```

## 3. 자율적으로 생각하고 판단하는 에이전트 구현
- 위 코드들은 자율적 판단보다는 검색 도구와 메모리 도구를 활용해 정해진 복잡한 작업을 수행하는 모델이었음
- 에이전트에게 스스로 생각해서 도구를 고를 수 있는 자율권을 줘보자
- 2개의 도구를 추가하여 코드 작성

```python
from datetime import datetime, date
```

```python
@tool
def get_current_time() :
    "현재 날짜 및 시간을 출력해주는 도구입니다."
    current_time = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    return current_time

@tool
def d_day_calculator(target_date_str:str) -> str:
    """특정 날짜(YYYY-MM-DD 형식)를 입력받아, 오늘 기준으로 며칠이 남았는지(D-day) 또는 며칠이 지났는지 계산해주는 도구입니다."""
    try :
        # 오늘 날짜(date type)
        today = date.today()
        # LLM이 넘겨준 문자열 날짜를 파이썬 날짜 객체로 변환(파싱)
        target_date= date.fromisoformat(target_date_str)
        # 날짜 차이를 일수로 계산
        days = (target_date - today).days

        if days > 0 :
            return f'{days}일 남았습니다. (D-{days})'
        elif days < 0 :
            return f'{days}일 지났습니다. (D+{days})'
        else:
            return '당일입니다! (D-Day)'
        
    except ValueError:
        return '날짜 형식이 올바르지 않습니다. "YYYY-MM-DD"형식으로 입력해주세요.'
    
```

```python
# create_tool_calling_agent : `프롬프트, LLM, 도구를 하나로 통합`하여 운용하는 에이전트 클래스
# AgentExecutor : 실제로 도구를 실행하고, 결과를 다시 LLM에게 가져다주는 실행기
from langchain_classic.agents import create_tool_calling_agent, AgentExecutor
```

```python
# 1. 도구 리스트
my_tools = [search_videos_tool, get_current_time, d_day_calculator]

# 2. LLM 로드
llm = ChatOpenAI(
    model='gpt-4o',
    api_key=MY_API_KEY,
    temperature=0.1   # 도구 선택의 일관성을 위해 낮게 설정
)

# 3. 에이전트용 프롬프트 (페르소나, 메모리, 스크래치패드)
prompt = ChatPromptTemplate.from_messages([
    ('system', """당신은 사용자의 일상을 돕는 똑똑한 AI비서입니다. 사용자의 요청을 분석하여 다음 도구들을 활용하세요:
     1. 동영상 검색도구(search_videos_tool): 유튜브 등 영상 컨텐츠나 음악을 찾을 때 사용
     2. 현재 날짜 및 시간 계산 도구 (get_current_time)
     3. D-day 계산 도구 (d_day_calcurator): 특정 기념일, 행사 또는 날짜가 며칠 남았는지/지났는지 물어볼 때 사용

     *주의*: 날짜 계산 도구를 사용할 때 올해의 연도가 명시되어 있지 않다면, 현재 년도(2026년)를 기준으로 삼으세요.
     도구의 결과를 바탕으로 상요자에게 친절하고 완성도 있는 문장으로 답변하세요.
     """),
    MessagesPlaceholder(variable_name='chat_history'),
    ('human', '{question}'),

    # agent_scratchpad : LLM이 사용한 도구를 기록하고 관리하는 공간 (약속된 예약어)
     # create_tool_calling_agent 를 사용할 때 'agent_scratchpad'를 반드시 설정해야 함
     # 또 프롬프트 템플릿 내에서 마지막 순서로 위치시켜야 함
    MessagesPlaceholder(variable_name='agent_scratchpad')   # AI가 생각하는 공간
    ])

# 4. 에이전트 조립 (모델 + 도구 + 프롬프트)
my_agent = create_tool_calling_agent(llm, my_tools, prompt)


# 5. 실행기(Executor) 생성
agent_executor = AgentExecutor(
    agent=my_agent,
    tools=my_tools,
    verbose=True,       # 에이전트 행동 단계 상세 과정 출력
    # <아래 두 매개변수는 에이전트용 안전장치>
    max_iteration=5,     # 사용자의 한 질문에 최대 5번까지만 도구 사용 (무한루프로 비용 폭발 방지)
    handle_parsing_errors=True   # 에이전트가 출력형식(Json)을 헷갈려 에러가 날 경우 자동 재시동 (직접 문자열로 재시도 프롬프트 넣어도 됨)
)


#### 6-1. 세션ID(여러 사용자)에 맞추어 DB 메모리를 연동하는 함수
def get_session_history_db(session_id:str) :
    DB_USER = 'root'    
    DB_PASS = '1234'  
    DB_HOST= '127.0.0.1'   
    DB_PORT= '3306'    
    DB_NAME='myAgent'  

    connection_str=f"mysql+pymysql://{DB_USER}:{DB_PASS}@{DB_HOST}:{DB_PORT}/{DB_NAME}?charset=utf8mb4"

    return SQLChatMessageHistory(
        session_id=session_id,          
        connection=connection_str,   
        table_name='chat_history_log'
        )

#### 7. 대화 메시지 기록 객체에 체인 융합
agent_with_history = RunnableWithMessageHistory(
    agent_executor,           # 실행기 객체               
    get_session_history_db,      
    input_messages_key = 'question', 
    history_messages_key='chat_history'
)

print("다중 유저 에이전트 비서가 준비되었습니다. (종료하려면 '종료' 입력)")



current_user = input('접속할 ID를 입력하세요 (예: user_1, user2) : ')
while True :
    print('-'*50)
    # 대화를 진행할 사용자(세션) 선택
    

    if current_user in '종료' :
        print('시스템을 종료합니다.')
        break
    
    user_input = input(f'[{current_user}의 질문을 입력하세요: ]')

    if user_input in '종료' :
        print('시스템을 종료합니다.')
        break

    # 선택한 id로 config 동적 생성  (configuration 환경설정)
    config = {'configurable': {'session_id':current_user}}

    # 융합된 에이전트 실행
    print(f'\n [{current_user}]의 과거 기록을 바탕으로 에이전트가 생각 중입니다...')
    response = agent_with_history.invoke(
        {'question': user_input},
        config=config
    )

    print(f'\n========= [{current_user}]님을 위한 최종 답변 ==========')
    print(response['output'])
    
```

```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain.tools import tool
from duckduckgo_search import DDGS
from datetime import datetime, date
from langchain_classic.agents import create_tool_calling_agent, AgentExecutor

load_dotenv()
MY_API_KEY = os.getenv('OPENAI_API_KEY')

@tool
def company_search_tool(keyword) :
    """
    기업의 최신 채용 공고, 공식 기술 블로그 등을 검색하는 도구입니다.
    [주의사항] 검색어(query)를 만들 때 단순히 기업 이름만 넣지 마세요!
    반드시 "네이버 채용 공고", "카카오 데이터 분석 인재영입" 처럼 
    '기업명 + 직무 + 채용/인재영입' 키워드를 구체적으로 조합해서 검색어를 생성하세요.
    """

    try :
        search_keyword = f"{keyword} 채용 OR 인재영입 OR 기술블로그"
        with DDGS() as ddgs :
            results = ddgs.text(
                keywords=search_keyword,
                region='kr-kr',
                timelimit='y',
                max_results=5
            )
        if not results :
            return f'{search_keyword}에 대한 검색 결과가 없습니다..'
        
        output_list = []

        # 반환 형태를 보기 좋게 설정
        for i in results :
            title = i.get('title', '제목 없음')
            body = i.get('body', '상세 내용 없음')
            link = i.get('href', '링크 없음')
            output_list.append(f'[제목]: {title}\n[요약]: {body}\n[링크]: {link}')

        return '\n\n'.join(output_list)
    
    except Exception as e : 
        return f'기업 동향 검색 중 오류 발생: {str(e)}, 오류를 해결하여 다시 시도하세요'


@tool
def d_day_calculator_tool(target_date:str) :
    """특정 날짜(YYYY-MM-DD 형식)를 입력받아, 오늘 기준으로 며칠이 남았는지(D-day) 또는 며칠이 지났는지 계산해주는 도구입니다."""
    try :
        today = date.today()
        target_date= date.fromisoformat(target_date)
        days = (target_date - today).days

        if days > 0 :
            return f'채용일이 {days}일 남았습니다. (D-{days})'
        elif days < 0 :
            return f'채용일이 {days}일 지났습니다. (D+{days})'
        else:
            return '당일입니다! (D-Day)'
        
    except ValueError:
        return '날짜 형식이 올바르지 않습니다. "YYYY-MM-DD"형식으로 입력해주세요.'


@tool
def video_search_tool(query: str) :
    """유튜브를 포함한 인터넷 동영상 자료를 검색할 때 사용합니다.
    사용자로부터 검색어(query)를 받아 관련된 동영상 제목과 링크를 반환합니다."""
    try :
        with DDGS() as ddgs :
            results = ddgs.videos(
                keywords=query,
                region='kr-kr',
                timelimit=None,
                max_results=10
            )
        if not results :
            return f'{query}에 대한 검색 결과가 없습니다..'
        
        output_list = []

        # 반환 형태를 보기 좋게 설정
        for i in results :
            title = i.get('title', '제목 없음')
            link = i.get('content', '링크 없음')
            output_list.append(f'[제목]: {title}\n[링크]: {link}')

        return '\n\n'.join(output_list)
    
    except Exception as e : 
        return f'영상 검색 중 오류 발생: {str(e)}, 오류를 해결하여 다시 시도하세요'


# ==========================================

tools = [company_search_tool, d_day_calculator_tool, video_search_tool]

llm = ChatOpenAI(model='gpt-4o', api_key=MY_API_KEY, temperature=0.1)

prompt = ChatPromptTemplate.from_messages([
    ('system', """당신은 사용자의 취업을 돕는 똑똑한 AI비서입니다. 사용자의 요청을 분석하여 다음 도구들을 활용해 답변하세요.
    1. 기업 검색도구(company_search_tool): 사용자가 기업을 언급하면 먼저 기업 동향 검색 도구를 사용할 것.
    2. 날짜 계산 도구(d_day_calculator_tool): 날짜가 주어지면 D-day 계산 도구를 사용할 것.
    3. 동영상 검색 도구(video_search_tool): 동영상을 검색할 때 사용.
    
     *주의*: 날짜 계산 도구를 사용할 때 올해의 연도가 명시되어 있지 않다면, 현재 년도(2026년)를 기준으로 삼으세요.
     답변의 마무리는 항상 면접이나 서류 준비에 도움이 되는 영상 추천으로 끝내세요.
     """),
    MessagesPlaceholder(variable_name='chat_history'),
    ('human', '{question}'),
    MessagesPlaceholder(variable_name='agent_scratchpad')   # AI가 생각하는 공간
])

# 에이전트 조립
career_agent = create_tool_calling_agent(llm, tools, prompt)


# 5. 실행기(Executor) 생성
agent_executor = AgentExecutor(
    agent=career_agent,
    tools=tools,
    verbose=True,
    max_iteration=5,    
    handle_parsing_errors=True 
)


#### 6-1. 세션ID(여러 사용자)에 맞추어 DB 메모리를 연동하는 함수
def get_session_history_db(session_id:str) :
    DB_USER = 'root'    
    DB_PASS = '1234'  
    DB_HOST= '127.0.0.1'   
    DB_PORT= '3306'    
    DB_NAME='myAgent'  

    connection_str=f"mysql+pymysql://{DB_USER}:{DB_PASS}@{DB_HOST}:{DB_PORT}/{DB_NAME}?charset=utf8mb4"

    return SQLChatMessageHistory(
        session_id=session_id,          
        connection=connection_str,   
        table_name='career_assistant_history'
        )

#### 7. 대화 메시지 기록 객체에 체인 융합
agent_with_history = RunnableWithMessageHistory(
    agent_executor,           # 실행기 객체               
    get_session_history_db,      
    input_messages_key = 'question', 
    history_messages_key='chat_history'
)

# =======================================================================

print("스마트 취업 에이전트 비서가 준비되었습니다. (종료하려면 '종료' 입력)")

current_user = input('접속할 ID를 입력하세요 (예: user_1, user_2, user_3) : ')
while True :
    print('-'*50)
    
    if current_user in '종료' :
        print('시스템을 종료합니다.')
        break
    
    user_input = input(f'[{current_user}의 질문을 입력하세요: ]')

    if user_input in '종료' :
        print('시스템을 종료합니다.')
        break

    # 선택한 id로 config 동적 생성  (configuration 환경설정)
    config = {'configurable': {'session_id':current_user}}

    # 융합된 에이전트 실행
    print(f'\n [{current_user}]의 과거 기록을 바탕으로 에이전트가 생각 중입니다...')
    response = agent_with_history.invoke(
        {'question': user_input},
        config=config
    )

    print(f'\n========= [{current_user}]님을 위한 최종 답변 ==========')
    print(response['output'])
```
