### 환경 변수 설정
- 파이썬 작업 폴더 경로에 txt 파일을 하나 생성하고 확장자명을 포함해서 .env로 파일 변경
- .env 파일은 메모장으로 열기(그냥 열면 비밀번호를 요구함)
- .env 파일을 메모장으로 열어서 내부에 **OPENAI_API_KEY=본인KEY** 넣어주기

```python
import os

# 환경변수 컨트롤 클래스 (코드 내부에 민감한 정보를 넣지 않고 따로 빼서 관리)
from dotenv import load_dotenv
from openai import OpenAI
```

### API Key 설정
- API key는 요금과 직결되는 민감 정보이므로 환경변수로 감춰서 사용

```python
load_dotenv()

#.env 파일 내의 OPEN_API_KEY 변수에 대한 값을 가져오기
MY_API_KEY = os.getenv("OPENAI_API_KEY")
```

```python
# OpenAI 객체 생성 - 키가 있어야 생성 가능
client = OpenAI(api_key=MY_API_KEY)
```

### GPT 채팅 API 불러와서 일반적인 질의

```python
question = "안녕? 오늘은 즐거운 목요일이야!"

# chat.completions.create : OpenAI의 chat API(대화)를 생성
response = client.chat.completions.create(
    model='gpt-3.5-turbo',   # 모델명은 공식 페이지의 모델명 중 하나로 사용
    messages=[{'role':'user', 'content':question}],
    # messages : 질의 정보를 설정하는 매개변수
     # role: 메시지의 역할을 설정하며 'system', 'user', 'assitant' 중 하나로 지정
     # content: 질의 내용 입력
    max_tokens=200,  # 생성할 응답의 최대 토큰 수
    temperature=0  # temperature: 모델 응답의 일관성 및 창의성 정도를 조정해주는 매개변수 (디폴트 1.0)
                    # (0~2사이 실수값으로 0은 항상 일관된 대답, 2는 매우 창의적인 대답)
                    # 보통 코딩/데이터분석은 0~0.2, 사실기반 정보 전달은 0.2~0.5, 일반 대화는 0.5~0.7, 창의적 글쓰기는 0.7~1.2
)
```

```python
response
```

### ChatCompletion 객체에는 사용자의 요청과 모델의 응답에 대한 정보가 담겨있음
- content에 실제 자연어 응답이 있으며, role에는 각 메시지의 역할, model 사용된 모델명, usage에는 사용된 토큰량에 대한 정보
- 모델이 응답시 role은 항상 assistant로 출력되며, 우리가 질의할 땐 user로 설정됨
- usage에 prompt_tokens는 입력 토큰 수, completion_tokens는 출력 토큰(응답) 수

채팅 API를 사용하고 있기 때문에, 상위모델일지라도 시각화 자료가 출력되는 것이 아니라, 텍스트로만 출력됨

ChatGPT 웹 서비스는 차트나 이미지를 출력하기 위한 다른 종류의 API도 같이 동작하고 있어, 요청에 맞는 응답을 해줌

```python
question = '지난 1년간의 판매 데이터를 시각화하기 위한 막대차트 생성해줘'

response = client.chat.completions.create(
    model='gpt-5-mini',    # gpt-5는 답변하기 전 혼자서 논리적인 생각을 거치고 검증하는 단계가 있음
                           # 사용자 질문에 대해 더 오래 생각하는 thinking모드와 빠른 응답의 fast 모드가 있음
                           # 질문에 따라 모델이 자율적으로 판단하여 응답함
    messages=[{'role':'user', 'content':question}],
    reasoning_effort='medium'  # 자율적인 생각을 방해할 수 없도록 temperature 매개변수가 없고,
                             # 생각의 깊이를 조절하는 reasoning_effort가 있음
                             # low: 빠르고 가볍게 응답. 비용 절감, 단순 작업
                             # medium: 적당히 고민해서 응답 (디폴트)
                             # high: 시간이 들어도 완벽한 논리로 대답 (복잡 코딩, 팩트 체크)
)

print(response.choices[0].message.content)
```

### 모델 로드 및 질의

```python
from langchain_openai import ChatOpenAI
```

## 프롬프트 템플릿 활용
- 질의시 규격화된 템플릿을 활용하여 좀 더 체계적인 구성으로 질의 가능
- 모듈화, 코드 재사용성, 일관성(버전관리 및 테스트 용이) 향상

```python
# PromptTemplate : 언어 모델에 대한 프롬프트(질의 및 응답) 관리를 용이하게 해주는 클래스
from langchain_core.prompts import PromptTemplate
```

```python
# from_template : 문자열 포맷팅을 기본적으로 활용하여 프롬프트 설정해주는 함수
pt = PromptTemplate.from_template('{context} 중에서 {topic}은?')

# 템플릿 내부에 들어갈 각각의 문자열 지정
prompt = pt.format(context='직업',
                   topic='도전적인 직업'
                   )

print(prompt)
```

```python
def get_answer(context, topic) : 
    pt = PromptTemplate.from_template('{context} 중에서 {topic}은?')

    prompt = pt.format(context=context, topic=topic)
    answer = chat_model.invoke(prompt)

    return f'[질의]: {prompt}, [응답]: {answer.content.strip()}'
```

```python
print(get_answer('직업', '육체적으로 힘든 직업'))
print(get_answer('직업', '정신적으로 힘든 직업'))
print(get_answer('직업', '연봉이 가장 높은 직업'))
print(get_answer('직업', '만족도가 가장 높은 직업'))
```
