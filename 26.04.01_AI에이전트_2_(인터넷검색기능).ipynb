#### 1) 비디오 검색 도구 선언

```python
# 데코레이션 형태로 에이전트가 사용할 수 있는 도구로 설정
 # @tool 이라 명시하면 langchain이 함수명, 설명(Docstring), 매개변수 타입 등을 읽어 JSON 설명서를 직접 만들어주고 도구 객체로서 동작하게 설정
 # 또한 OpenAI GPT 모델 뿐만 아니라 구글이나 엔트로픽 등 다른 모델도 @tool로 설정 가능
@tool
def search_videos_tool(query: str) -> str :
    # 에이전트가 사용할 도구에 대한 설명서 (docstring)
    # langchain의 @tool 사용시에는 docstring이 필수적으로 들어가야 함
    # (OpenAI에서 tool을 사용할 때 description의 역할)
    """유튜브를 포함한 인터넷 동영상 자료를 검색할 때 사용합니다.
    사용자로부터 검색어(query)를 받아 관련된 동영상 제목과 링크를 반환합니다."""

    try :
        # DDGS()는 덕덕고 서버와 통신하는데 with문이 없다면 원하는 작업이 끝나도 통신이 끊어지지 않고
        # 메모리와 연결포트를 점유하기 때문에 작업이 끝날때 같이 연결도 끊어질 수 있도록 with문으로 작성
        with DDGS() as ddgs :
            results = ddgs.videos(
                keywords=query,
                region='kr-kr',
                timelimit=None,
                max_results=10
            )
            # <DDGS 객체의 다른 메소드>
             # text(): 일반 웹 검색
             # images()
             # videos()
             # news()
             # maps(): 지도/장소 검색

        if not results :
            return f'{query}에 대한 검색 결과가 없습니다..'
        
        output_list = []

        # 반환 형태를 보기 좋게 설정
        for i in results :
            title = i.get('title', '제목 없음')
            link = i.get('content', '링크 없음')
            output_list.append(f'[제목]: {title}\n[링크]: {link}')

        return '\n\n'.join(output_list)
    
    except Exception as e : 
        return f'영상 검색 중 오류 발생: {str(e)}, 오류를 해결하여 다시 시도하세요'
```

```python
searh5 = search_videos_tool.invoke('화사 신곡')
print(searh5)
```

- `RatelimitException` 에러는 덕덕고 서버가 사람이 아닌 봇이라고 판단해 나오는 에러
- 덕덕고는 정보 보안을 중요시하는 서버이므로 크롤링 시 종종 IP가 잠시 차단되기도 함 (보통 10~15분)
- 크롤링 로직이 들어간 검색 라이브러리를 사용한다면 높은 안정성이 필요한 프로젝트의 경우 정식 결제하고 검색 API 사용하기

#### 2) DDGS 객체의 다른 메소드 활용법

```python
# 텍스트 검색 예시
with DDGS() as ddgs :
    print("=== 텍스트 검색 결과 ===")
    results = ddgs.text(
        keywords='화사 신곡',
        region='kr-kr',
        timelimit=None,
        max_results=5
    )

    # 확인용 코드
    print(results)
    print()

    for i, txt in enumerate(results, 1) :
        print(f'{i}. 제목:{txt['title']}')
        print(f'   링크: {txt['href']}')         # href 링크
        print(f'   내용: {txt['body'][:50]}...')  # body 내용
        print()
```

```python
# 이미지 검색 예시
with DDGS() as ddgs :
    print("=== 이미지 검색 결과 ===")
    results = ddgs.images(
        keywords='벚꽃',
        region='kr-kr',
        timelimit=None,
        max_results=3
    )

    for i, img in enumerate(results, 1):
        print(f"{i}. 이미지 URL: {img['image']}")    # 이미지 링크
        print(f'   링크: {img['source']}')         # 출처
        print()
```
