코랩 클라우드 플랫폼에 이미지 데이터를 업로드하면 시간이 많이 걸림

어차피 이미지를 전부 array로 변환할 거니까, 로컬에서 array로 변환하고 해당 데이터를 코랩으로 옮겨서 사용하자~!  

```python
import os        # Operating System 파일이나 폴더 처리 라이브러리
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from PIL import Image
```

```python
hedgehog_dir = '고슴도치/'
hamster_dir = '햄스터/'
meerkat_dir = '미어캣/'
```

```python
# listdir : 지정된 폴더 내 모든 파일의 명칭을 리스트로 반환
hedgehog_fnames = os.listdir(hedgehog_dir)
hamster_fnames = os.listdir(hamster_dir)
meerkat_fnames = os.listdir(meerkat_dir)
```

```python
hedgehog_fnames
```

```python
# path.join : 여러 인자를 넣고 하나의 경로로 만들어주는 함수
test_path = os.path.join(hedgehog_dir, hedgehog_fnames[50])
test_path
```

```python
# 폴더 경로, 파일명, 이미지 사이즈를 매개변수로 받아 / 이미지를 로드한 후 배열로 변환하는 함수

def load_images(folder_path, file_names, img_size_shape=(224,224)) : 
    images= []
    for i in file_names : 
        # 경로 합치기 (폴더명 + 파일명)
        path = os.path.join(folder_path, i)
        # resize : 배열의 사이즈 변경
        # convert('RGB') : 컬러 이미지로 받기
        img = Image.open(path).resize(img_size_shape).convert('RGB')

        # 이미지 불러와서 array로 변경하고, 빈리스트에 넣어주기
        images.append(np.array(img))
        
    return np.array(images)        # 반환값도 배열로 변환
```

>img_size_shape라는 매개변수로 뺐을 때: 함수를 호출할 때 load_images(folder, files, (128, 128)) 이렇게 숫자만 바꿔주면 끝!!
>
>함수 내부를 건드릴 필요가 없지!

```python
hedgehog = load_images(hedgehog_dir, hedgehog_fnames)
hamster = load_images(hamster_dir, hamster_fnames)
meerkat = load_images(meerkat_dir, meerkat_fnames)
```

```python
print(hedgehog.shape)
print(hamster.shape)
print(meerkat.shape)

# 맨 마지막 3의 의미는 컬러이미지라는 뜻 (R,G,B) 
```

```python
# 문제 데이터 생성
X = np.concatenate([hedgehog, hamster, meerkat])
X.shape
```

```python
# 정답 데이터 생성
y = np.array([0]*414 + [1]*421 + [2]*390)
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    random_state=22,
    stratify=y
)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# npz(Numpy Zip) : 배열 데이터를 압축한 파일 형식
np.savez_compressed('data/animals.npz',
                   X_train = X_train,
                   X_test = X_test,
                   y_train = y_train,
                   y_test = y_test
                   )
```
