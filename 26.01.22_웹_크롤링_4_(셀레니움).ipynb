### XPATH 절대 경로
- XPATH는 원하는 하나의 요소에 정확히 접근하기 위한 절대 경로를 가져오는 명령
- 하나의 요소에 접근할 때 사용하면 편리하나, 여러개의 요소에 접근해야 하거나 웹 페이지가 유동적으로 변화하는환경에서는 효율이 떨어짐
- 따라서 변동이 적은 개별 고정 요소를 추출할 때 사용하거나, 요소에 ID, Class와 같은 구분자가 제대로 붙어있지 않을 때 사용하면 좋음
- HTML 코드에서 우클릭 후 copy - copy XPATH로 복사해서 사용 

```python
# 돋보기 버튼 클릭시 XPATH 사용
btn = driver.find_element(By.XPATH, '//*[@id="search-btn"]')
btn.click()
```

## try-except 예외 처리
- 동적 크롤링은 변수 상황이 많기 때문에 예외처리를 정말 많이 사용하게 됨
  - **try**: 일반적으로 동작하는 코드
  - **except**: try문이 동작하지 않는 예외의 경우(에러 발생시) 작동되는 코드

```python
# try문: 일반적으로 동작하는 코드
try:
    # 더보기 버튼이 있으면 계속 반복 
    while True:
        btn = driver.find_element(By.CSS_SELECTOR, 'a.c_05')
        btn.click()
        time.sleep(2)
        
# except문: 에러로 try문이 동작하지 않는 예외의 경우
except:
    print("더이상 더보기 버튼이 없네요 ㅠ_ㅠ 그래도 코드는 이어서 동작합니다~!")

# 더보기 버튼을 다 누르고 난 뒤 (HTML요소를 다 받아온 뒤) 메뉴와 가격을 추출
# HTML 요소 모두 받은 뒤이므로, time.sleep을 줄 필요가 없음


titles = driver.find_elements(By.CSS_SELECTOR, 'h4.h.fz_03')
prices = driver.find_elements(By.CSS_SELECTOR, 'div.item-price > strong')
titles[-1].text, prices[-1].text

titles_list, prices_list = [], []

for i in tqdm(range(len(titles))):
    titles_list.append(titles[i].text)
    prices_list.append(prices[i].text + '원')

print('크롤링 완료!')
```

```python
len(titles_list), len(prices_list)
```

### 두개의 리스트의 요소들을 1:1로 묶어서 딕셔너리로 만들기

```python
# zip: 길이가 동일한 두 개의 집합의 요소들을 1:1로 대응시켜 튜플로 묶어주는 함수
# zip 함수는 묶여진 데이터가 바로 표시되진 않아서 변수로 출력시켜도 바로 확인은 불가

temp = zip(titles_list, prices_list)
temp
```

```python
# zip 객체를 딕셔너리로 형 변환

title_price_dict = dict(zip(titles_list, prices_list))
title_price_dict
```

# 여태까지 작성한 코드를 한 셀로 동작시키기
- 만약 웹 크롤러를 개발하고 실행파일을 만들어서 배포한다면 한 셀에 시작부터 끝까지 코드가 모두 들어가 있어야 함!

```python

#     from selenium import webdriver as wb
#     from selenium.webdriver.common.by import By
#     from selenium.webdriver.common.keys import Keys
#     from tqdm import tqdm
#     import pandas as pd
#     import time

driver = wb.Chrome()
driver.get("https://www.hsd.co.kr/menu/menu_list")
time.sleep(1)
driver.maximize_window()
time.sleep(1)

try:
    while True:
        btn = driver.find_element(By.CSS_SELECTOR, 'a.c_05')
        btn.click()
        time.sleep(2)
except:
    print("더이상 더보기 버튼이 없네요 ㅠ_ㅠ 그래도 코드는 이어서 동작합니다~!")

titles = driver.find_elements(By.CSS_SELECTOR, 'h4.h.fz_03')
prices = driver.find_elements(By.CSS_SELECTOR, 'div.item-price > strong')

# titles_list = [i.text.strip() for i in titles]
# prices_list = [i.text.strip() for i in prices]

titles_list, prices_list = [], []

for i in tqdm(range(len(titles))):
    titles_list.append(titles[i].text)
    prices_list.append(prices[i].text + '원')

print('크롤링 완료!')
print('메뉴 개수:', len(titles_list))
print('가격 개수:', len(prices_list))

hansot_dic = {"메뉴명": titles_list, "가격": prices_list}

hansot_df = pd.DataFrame(hansot_dic, index=range(1,len(titles_list)+1))
hansot_df.index.name = 'No.'
hansot_df
```
