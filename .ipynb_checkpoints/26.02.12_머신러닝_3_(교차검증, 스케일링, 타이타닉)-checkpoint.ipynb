# **iris 데이터를 활용하여 스케일링을 진행~!**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
iris_data = load_iris()
iris_data

# data : 실제 데이터
# target : 정답 데이터
# feature_names : 컬럼명
```

```python
type(iris_data)

# Bunch : sklearn에서 내부적으로 사용하는 딕셔너리
#         딕셔너리처럼 '[]'로 접근할 수도 있고, '.'으로 접근할 수도 있음
```

```python
iris_df = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
iris_df.head()
```

```python
# bar차트로 시각화
plt.figure(figsize=(15, 5))
plt.bar(iris_df.index, iris_df['sepal length (cm)'])
plt.title('before data scaling')
plt.show()
```

```python
# 히스토그램 (수치형 데이터에서 구간별 분포 보려고~~)
plt.hist(iris_df['sepal length (cm)'], bins=30)
plt.title('before data scaling')
plt.show()
```

## **Standard Scaler**

```python
scaler_SS = StandardScaler()     # 스케일러 객체 생성
scaler_SS.fit(iris_df)       # iris 데이터의 범위로 스케일러 학습
iris_SS = scaler_SS.transform(iris_df)    # 학습된 스케일러로 실제로 값 변환
```

```python
iris_SS
```

```python
iris_SS[:, 0]    # 모든 행의 첫번째 열 ('sepal length (cm)')
```

```python
# bar차트로 시각화
plt.figure(figsize=(15, 5))
plt.bar(iris_df.index, iris_SS[:, 0])
plt.title('after SS scaling')
plt.show()
```

```python
# 히스토그램 (수치형 데이터에서 구간별 분포 보려고~~)
plt.hist(iris_SS[:, 0], bins=30)
plt.title('after SS scaling')
plt.show()

# 스케일링 전후로 데이터의 분포와 비율은 동일!!!
```

## Standard Scaler로 변환된 값들의 평균과 분산 확인하기

```python
# 스케일링된 값으로 DF 생성 
iris_df_SS = pd.DataFrame(iris_SS, columns=iris_data.feature_names)
iris_df_SS.head()
```

```python
print('<원본 데이터 특성별 평균>')
print(iris_df.mean())
print('-'*30)
print('<원본 데이터 특성별 분산>')
print(iris_df.var())
```

```python
print('<스케일링 데이터 특성별 평균>')
print(iris_df_SS.mean())
print('-'*30)
print('<스케일링 데이터 특성별 분산>')
print(iris_df_SS.var())
```

# **교차검증 (Cross Validation)**

```python
X_train, X_test, y_train, y_test = train_test_split(
    iris_data.data, iris_data.target,
    test_size=0.3,
    random_state=42,
    stratify=iris_data.target
)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors=10)
```

```python
knn_model.fit(X_train, y_train)
knn_model.score(X_train, y_train)
```

```python
# cross_val_score(모델, 문제, 정답, cv=나눌개수)
result = cross_val_score(
    knn_model,
    X_train, y_train,
    cv=5        # 나눠줄 데이터셋 개수
    ) 
```

```python
result

# cv를 5로 설정하면 train 데이터의 총 개수인 105개를 5개의 데이터셋으로 분리시킨다는 뜻~
# 그 중 4개 셋트로 학습하고 나머지 1개의 셋트로 평가하는 것을 바꿔가며 5번 반복함
# 교차검증은 최종평가 전, train데이터만으로 사전에 데이터 및 모델을 검증하는 용도로 사용 
```

## 1) 결측치 처리

#### **1. Age 컬럼의 결측치 채우기**
- 나이는 전체 평균으로 넣기엔 범위가 너무 넓고, 데이터 분포가 치우쳐 있어서 대표값으로 생각하기 애매
- Age 컬럼과 상관관계가 높은 다른 컬럼들의 데이터를 이용해 결측치를 좀 더 세분화된 값으로 채워주자

```python
# corr : 상관계수 출력 함수
# 상관계수는 각 특성들 간의 상관관계의 척도를 수치로 판단할 수 있고, 절대값이 클수록 관계의 영향도가 높아짐
# 일반적으로 절대값이 0.7이상이면 강한 상관관계, 0.3이하면 약한 상관관계
# 범위는 -1(반비례) ~ 1(정비례)

train.corr(numeric_only=True)       # numeric_only=True 수치데이터만 고려
```

```python
# https://matplotlib.org/stable/users/explain/colors/colormaps.html
# 상관계수를 시각화~~!

plt.figure(figsize=(7, 5))
sns.heatmap(data=train.corr(numeric_only=True),
           annot=True,        # annotation : 각 셀 내부 값 표시
           fmt='.2f',         # 소수 2번째자리까지 표시
           annot_kws={'size':12},  # 수치 사이즈 조절
           cmap='Blues'       # 히트맵 색상
           );
```

```python
test.corr(numeric_only=True)
```

### Age와 상관관계가 높은 Pclass 컬럼을 결측치 채우는데 활용하자
- 생존에 관련이 많았던 성별 컬럼도 함께 활용하기 

```python
# pivot_table : 컬럼들의 집계된 정보를 출력할 수 있는 테이블 형태의 표 생성

pt1 = train.pivot_table(values='Age',           # 데이터로 사용될 컬럼
                       index=['Pclass', 'Sex'],   # 인덱스로 사용될 컬럼 (multi index)
                       aggfunc='mean'         # 집계 함수 설정
                       )
pt1
```

```python
# 멀티 인덱싱
pt1.loc[1, 'female']
```

```python
# isna : 결측치 (결측이면 True)
pd.isna(train['Age'])        # 수치형, 문자형 데이터 모두 판단

# train['Age'].isnull()      # 수치형, 문자형 데이터 모두 판단
# np.isnan(train['Age'])     # 수치형만
```

```python
# 불리언 인덱싱으로 결측치인 행만 출력
train[train['Age'].isnull() == True]
```

```python
pt1.loc[3, 'male'].iloc[0]   # 시리즈여서 iloc로 단일값만 뽑아낸 것
round(pt1.loc[3, 'male'].iloc[0])
```

```python
# 상황에 맞게 Age 채워주는 함수 선언

def fill_age(df) :          # df에는 train or test 들어갈 예쩡
    if pd.isna(df['Age']):       # 결측치인 경우, 
        return round(pt1.loc[df['Pclass'], df['Sex']].iloc[0])  # 결측치가 있는 행의 Pclass,Sex 컬럼 반영하여 멀티 인덱싱한 값을 리턴
    else:
        return int(df['Age'])  # 자료형만 정수로 변환하여 반환
```

```python
# apply : 데이터프레임(혹은 시리즈)에서 행 또는 열에 한 번에 복잡한 계산을 적용할 때 사용
train['Age'] = train.apply(fill_age, axis=1)

# 동작 로직 -> fill_age 함수의 매개변수로 train을 넣고, 컬럼(열) 단위로 해당 함수를 적용
```

```python
train['Age']

# 결측치가 아닌 데이터들은 정수형태로 변환, 결측치인 데이터는 피벗테이블 정보가 들어감
```

```python
test['Age'] = test.apply(fill_age, axis=1)

# test 데이터 결측치를 채울 때에도, train 데이터를 기준으로 만든 pt1 적용

# 머신러닝 지도학습에서는 test 데이터에 대한 정보가 들어간 상태로 학습을 진행하면 간접 치팅
# test 데이터의 기타 정보들은 모른다고 가정하고 결측치에도 train 데이터를 기준으로 정보를 채워줌
```

#### **4. Cabin 결측치 채우기**

- but, 데이터를 보고 판단해야 하기 때문에 일단 결측치에 임의의 값을 넣어서 시각화 한 뒤, 삭제할지 넣을지 판단하기

#### SibSp - 함께 탑승한 형제자매, 배우자의 수
- 누군가와 함께 탑승-> 생존에 도움이 될 가능성이 있음
- 두 컬럼을 더해서 가족 숫자라는 새로운 컬럼을 생성해서 활용해보자!

```python
# 새 컬럼 생성 (혼자 탔으면 1)
train['Family_Size'] = train['Parch'] + train['SibSp'] + 1
test['Family_Size'] = test['Parch'] + test['SibSp'] + 1

train.head()
```

```python
plt.figure(figsize=(5,3))
sns.countplot(data=train,
             x='Family_Size',
             hue='Survived');
```

- 1명일 때 (혼자) 사망비율이 높고, 2~4명 일 때는 생존 비율이 높, 5명 이상이면 사망 비율이 높음

#### **Family_size가 1이면 Alone, 2~4면 Small, 5이상이면 Large로 범주화**

```python
bins_size = [0, 1, 4, 11]     # 시작점 미포함, 끝점 포함
labels_size = ['Alone', 'Small', 'Large']

# 범주화시킨 새 컬럼 Family_Group 생성
 # cut : 특정 데이터를 원하는 구간과 명칭으로 범주화 시켜주는함수 
 # bins : 범주의 경계 지정
 # label : 범주의 명칭 지정
train['Family_Group'] = pd.cut(train['Family_Size'],
                              bins=bins_size,
                              labels=labels_size
                              )

test['Family_Group'] = pd.cut(test['Family_Size'],
                              bins=bins_size,
                              labels=labels_size
                              )
```

```python
train['Family_Size'].head()
```

```python
train['Family_Group'].head()
```

```python
plt.figure(figsize=(5,3))
sns.countplot(data=train,
             x='Family_Group',
             hue='Survived');
```

- Family_Size로 보는 것보다 범주화시킨 Group으로 보는 것이 더 직관적
- 사람이 보기 편하면 기계도 이해하기 쉬울 가능성이 높음!

```python
# data에는 컬럼(Series)이 들어갈 예정
def split_title(data) :
    return data.split(',')[1].split('.')[0].strip()
```

```python
train['Title'] = train['Name'].apply(split_title)
test['Title'] = test['Name'].apply(split_title)

# apply 함수 앞에 1차원인 Series가 들어가기 때문에, axis 설정은 필요없음
```

- Rev는 목사님. 전원 사망.
- 데이터 수는 적지만 명확한 패턴을 보이므로, 정보를 활용할 예정
- Mr, Mrs, Miss, Master, Rev를 제외한 호칭들은 상대적으로 개수가 적으니 Other로 묶어주자 

```python
# unique 값을 그대로 복사해서 title  변수의 리스트에 넣기
# 'Don' 뒤로 보냄
title = ['Mr', 'Mrs', 'Miss', 'Master', 'Rev', 'Don', 'Dr', 'Mme', 'Ms',
       'Major', 'Lady', 'Sir', 'Mlle', 'Col', 'Capt', 'the Countess',
       'Jonkheer']

len(title)
```

```python
# 중요 호칭 5종류는 그대로 사용, 나머지는 Other로 변환하여 리스트 생성
convert_title = ['Mr', 'Mrs', 'Miss', 'Master', 'Rev'] + ['Other']*12
convert_title
```

```python
# zip : 두 개의 순서가 있는 집합의 요소들을 1:1로 매핑 후 딕셔너리로 변환
title_dict = dict(zip(title, convert_title))
title_dict
```

```python
# 기존 Title 컬럼을 map 함수를 사용하여 원하는 값으로 변환시킨 후 업데이트 
train['Title'] = train['Title'].map(title_dict)
```

## **모델 선택**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

# KFold : 교차검증 세부사항 설정
from sklearn.model_selection import cross_val_score, KFold
from tqdm import tqdm
```

## 1) **KNN 모델**

```python
knn_model = KNeighborsClassifier(n_neighbors=5)
```

```python
# n_splits : 데이터셋의 갯수
# shuffle : 데이터를 섞어서 분리할지 결정
# random_state는 셔플 했을 때 지정

kfold = KFold(n_splits=10,
             shuffle=True,
             random_state=42)
```

```python
# K-fold 교차검증 수행

result_KNN = cross_val_score(
    knn_model,
    X_train,
    y_train,
    cv=kfold
)

result_KNN
```

```python
result_KNN.mean()
```

```python
# 이웃 수에 따른 성능 그래프
train_score_list = []
for i in tqdm(range(1, 101, 1)):
    knn_model=KNeighborsClassifier(n_neighbors=i)   # 모델 객체 생성
    
    # 교차검증 실행
    result_KNN = cross_val_score(
        knn_model,
        X_train,
        y_train,
        cv=kfold
    )

    #평균 정확도 추출
    score_mean=result_KNN.mean()
    # 정확도값 리스트에 넣기
    train_score_list.append(score_mean)
```

```python
plt.figure(figsize=(15, 5))
plt.plot(range(1, 101, 1), train_score_list)
plt.xticks(range(1, 101, 2))   # 표시는 2칸에 하나씩
plt.grid()
plt.show()
```
