## 수집한 데이터 정제

```python
print(type(titles))
print(type(name))

# 겉으로 보기에는 자료형이 리스트로 보였지만 실제로는 BS내부에서 사용하는 ResultSet 이라는 객체임
# BS내부에서 사용하는 특수한 리스트로 pandas나 matplotlib에 바로 사용할 수 없어서 리스트로 변형해야 함~!
```

```python
titles_list, artist_list =[], [] # 빈 리스트 생성

# 반복문으로 컨텐츠(텍스트)만 추출하여 빈 리스트에 담기
for i in range(len(titles)): # titles나 name 모두 100개로 동일해서 뭘 사용해도 상관없음
    titles_list.append(titles[i].text.strip())
    artist_list.append(name[i].text.strip())

    # 보기 좋게 출력
    print(titles[i].text+ '-' +name[i].text)
```

```python
len(titles_list), len(artist_list)
```

```python
# 데이터프레임을 만들 딕셔너리 선언
melon_dic = {'노래제목':titles_list, '가수명':artist_list}
melon_dic
```

## 회사명

```python
# 1. 부모 태그와 선택자로 넓은 범위에서 접근 < .. ?????????????????
                        # 태그.클래스명     .클래스명
company = soup.select('div.col.company_nm .str_tit') # 태그가 각각 a, span이어서 문제였는데.. 클래스 이름이 str_tit으로 같아서 태그 생략하고 넣음
company
print(len(company))

company_list = []

for i in range(1, len(company)):
    company_list.append(company[i].text.strip())
print(company_list)
```

```python
# 2. 원하는 태그에 각각 접근하기
# 여러 태그가 복잡하게 얽혀 있을 경우, 조건문으로 상황에 맞게 태그 값을 지정해서 정보를 가져오기
c_list = []

for i in company:
    # a 태그에 값이 있다면, True로 인지함!
    target = i.select('a') or i.select('span')
        # select로 해도 되지만 리스트로 반환 받으므로 한 번 더 정제해야 하는데
        # select_one은 하나의 값만 반환해서 편하게 사용 가능
        # 어차피 우리가 원하는 회사명은 단일 값이므로 select_one으로 받는 게 더 효율적
        # target = i.select_one('a')
        
    # a 태그에 값이 없다면 span 태그를 확인    
    # elif i.select('span'):
        # target = i.select_one('span')
        
    if target:
        c_list.append(target.text.strip())
c_list
```

## iframe 태그
- **다른 서버의 웹 페이지 데이터나 리소스를 현재 서버의 웹 페이지에 포함**시켜 보여주는 HTML 태그
- 웹페이지에 보여줄 데이터가 많은 경우 한번에 접근하려면 그만큼 속도가 느려질 수 있는데, 이를 방지하거나 데이터를 따로 관리하기 용이하게 하기 위해 iframe 태그를 사용함!
- 우리가 수집하길 원하는 정보가 iframe 태그 내에 있다면, 기존 url로는 접근이 불가하고, 해당 데이터가 있는 원천 서버주소로 접근해야 함! (원하는 정보의 태그를 포함하는 내부 전체 \<HTML> 태그를 찾고 그 바로 위에 iframe 태그가 존재함 )

## 이제 1~10 p의 데이터를 모두 수집해보자!
- page의 경우 GET 방식의 웹사이트는 url에 정보가 표시되므로, 해당 정보가 어디있는지 파악해 볼 것

```python
# tqdm: 반복문의 진행상황을 로딩바로 확인하게 해주는 모듈 
from tqdm import tqdm
```

```python
https://www.10000recipe.com/recipe/list.html?order=reco&page=3
```

```python
titles_list, authors_list, views_list = [], [], []

for i in tqdm(range(1,11,1)) :
    # 1~10page에 반복적으로 접근
    url = f'https://www.10000recipe.com/recipe/list.html?order=reco&page={i}'
    res = requests.get(url, headers=U_A)
    soup = bs(res.text, 'lxml')

    titles = soup.select('div.common_sp_caption_tit.line2')
    authors = soup.select('div.common_sp_caption_rv_name > a')
    views = soup.select('span.common_sp_caption_buyer')

    for i in range(len(titles)) :
        titles_list.append(titles[i].text.strip())
        authors_list.append(authors[i].text.strip())
        views_list.append(views[i].text.strip()[4:])

recipe_dict = {'레시피명':titles_list, '작성자':authors_list, '조회수':views_list}
recipe_df = pd.DataFrame(recipe_dict, index=range(1,len(titles_list)+1))
recipe_df.index.name = 'No'
recipe_df
```

```python
filtered_df = recipe_df[recipe_df['레시피명'].str.contains('김치')]

# reset_index : 인덱스 재설정
filtered_df.reset_index(drop=True, inplace=True)
filtered_df.index = range(1, len(filtered_df)+1)
filtered_df
```

## 함수로 코드 통합하기

```python
# 매개변수는 검색할 최종페이지, 필터링 할 검색어 (디폴트로 빈 문자열)
# 반환할 자료형은 DataFrame을 명시
def search_recipe(end_page:int, search:str='') -> pd.DataFrame:
    
    titles_list, authors_list, views_list = [], [], []

    for i in tqdm(range(1, end_page+1, 1)) :
        # 1~10page에 반복적으로 접근
        url = f'https://www.10000recipe.com/recipe/list.html?order=reco&page={i}'
        res = requests.get(url, headers=U_A)
        soup = bs(res.text, 'lxml')
    
        titles = soup.select('div.common_sp_caption_tit.line2')
        authors = soup.select('div.common_sp_caption_rv_name > a')
        views = soup.select('span.common_sp_caption_buyer')
    
        for i in range(len(titles)) :
            titles_list.append(titles[i].text.strip())
            authors_list.append(authors[i].text.strip())
            views_list.append(views[i].text.strip()[4:])
    
    recipe_dict = {'레시피명':titles_list, '작성자':authors_list, '조회수':views_list}
    recipe_df = pd.DataFrame(recipe_dict, index=range(1,len(titles_list)+1))

    # 검색어가 존재할 경우!
    if search:
        # copy: 기존 객체를 복사
        filtered_df = recipe_df[recipe_df['레시피명'].str.contains(f'{search}')].copy()
    # 검색어 없으면
    else:
        # 기존 recipe_df를 그대로 사용
        filtered_df = recipe_df.copy()

    filtered_df.reset_index(drop=True, inplace=True)
    filtered_df.index = range(1, len(filtered_df)+1)
    return filtered_df
```

```python
search_recipe(2)
```
