### **input 태그**
- 사용자의 입력을 받기 위한 인터페이스를 제공하는 특수 태그
- 시작 태그만 있고 끝 태그는 없음
- 다양한 **속성**들을 가지고 있음
  - type: 입력 형식 지정
  - value: 서버로 전송할 값
  - disabled: 비활성 상태 (입력 불가)

```python
# 활성화, 비활성화 사이즈를 구분하기 위해 input 태그까지 접근
size = driver.find_elements(By.CSS_SELECTOR, '#optSizeSection > ul > li > input')

# get_attribute: 특정 요소의 속성값에 접근하는 함수
# 속성 값이 있으면 해당 값이 나오고, 없는 속성은 None 혹은 True로 반환
size[0].get_attribute('value')
```

```python
# print문으로 None 확인 가능
print(size[0].get_attribute('disabled'))
```

```python
# CSS_SELECTOR로 특정 태그의 속성에 접근하기
# 태그명[속성명=]
temp = driver.find_element(By.CSS_SELECTOR, 'label[for="size_220"]')
temp.text
```

```python
temp = driver.find_element(By.CSS_SELECTOR, 'label[title="230"]')
temp.text
```

```python
size = driver.find_element(By.XPATH, '//*[@id="size_300"]')
size.get_attribute('disabled')
```

```python
# 구매 가능 사이즈와 불가 사이즈 추출
enabled_size, disabled_size = [], []

for i in size:
    input_id = i.get_attribute('id') # size_220부터 5씩 증가되어서 input_id에 입력

    # 위의 input_id 값이 속성 for에 들어가는 문자열과 같으므로 포맷팅으로 코드 작성
    label = driver.find_element(By.CSS_SELECTOR, f'label[for="{input_id}"]')

    # 텍스트만 추출
    label_text = label.text.strip()

    # disabled 속성이 없다면 None이므로 아래 조건문에서 False로 동작함
    is_disabled = i.get_attribute('disabled')

    if is_disabled:
        disabled_size.append(label_text)
    else:
        enabled_size.append(label_text)

print('구매 가능 사이즈: ', enabled_size)
print('구매 불가 사이즈: ', disabled_size)
```

## 상위 4개 품목 정보 수집

```python
driver = wb.Chrome()
driver.get('https://www.nbkorea.com/product/featuredList.action?cateGrpCode=250100&cIdx=1385')
time.sleep(1)

imgs = driver.find_elements(By.CSS_SELECTOR, 'a.pro_area > img.img_goods')

titles_list, score_list, price_list, review_list = [], [], [], []

# while True:
for i in tqdm(range(4)):
    imgs = driver.find_elements(By.CSS_SELECTOR, 'a.pro_area > img.img_goods')
    imgs[i].click()
    time.sleep(2)

    # 크롤링
    title = driver.find_element(By.CSS_SELECTOR, 'h2.title')
    score = driver.find_element(By.CSS_SELECTOR, 'div.crema_product_reviews_score__container')
    price = driver.find_element(By.CSS_SELECTOR, 'strong#dispPrice')
    width = driver.find_element(By.CSS_SELECTOR, '#optWidthSection > ul > li > label')
    review = driver.find_element(By.CSS_SELECTOR, 'a.crema-product-reviews-count.crema-applied')

    titles_list.append(title.text)
    score_list.append(score.text)
    price_list.append(price.text)
    review_list.append(review.text)
    
    enabled_size, disabled_size = [], []

    size = driver.find_elements(By.CSS_SELECTOR, '#optSizeSection > ul > li > input')
    
    for i in size:
        input_id = i.get_attribute('id') # size_220부터 5씩 증가되어서 input_id에 입력
    
        label = driver.find_element(By.CSS_SELECTOR, f'label[for="{input_id}"]')
        label_text = label.text.strip()

        is_disabled = i.get_attribute('disabled')
    
        if is_disabled:
            disabled_size.append(label_text)
        else:
            enabled_size.append(label_text)
            
    print(title.text+" - "+score.text+" - "+width.text+" - "+review.text)
    print("구매가능 사이즈 :", enabled_size)

    driver.back()
    time.sleep(2)
```

# 처음부터 모든 코드를 한 셀에서 동작시키기
- 상위 12개 품목에 대해 순위를 매겨 데이터 프레임으로 출력하기까지...........

```python
driver = wb.Chrome()
driver.get('https://www.nbkorea.com/product/featuredList.action?cateGrpCode=250100&cIdx=1385')
time.sleep(1)
driver.maximize_window()
time.sleep(1)

# 각 상품 정보를 담아줄 빈 리스트 생성
title_list, mean_rating_list, enabled_size_list = [],[],[]
foot_width_list, cnt_reviews_list = [], []

try :
    for i in tqdm(range(12)) :
        imgs = driver.find_elements(By.CSS_SELECTOR, 'a.pro_area > img.img_goods')
        imgs[i].click()
        time.sleep(2)
    
        title = driver.find_element(By.CSS_SELECTOR, 'h2.title')
        
        mean_rating = driver.find_element(
            By.CSS_SELECTOR, 
            'div.crema_product_reviews_score__container')
        
        size = driver.find_elements(
            By.CSS_SELECTOR, 
            '#optSizeSection > ul > li > input')
        
        enabled_size, disabled_size = [], []
        for i in size :
            input_id = i.get_attribute('id')
            label = driver.find_element(By.CSS_SELECTOR, f'label[for="{input_id}"]')
            label_text = label.text.strip()
            
            is_disabled = i.get_attribute('disabled')
            if is_disabled :
                disabled_size.append(label_text)
            else :
                enabled_size.append(label_text)
                
        foot_width = driver.find_element(
            By.CSS_SELECTOR, 
            '#optWidthSection > ul > li > label')
    
        cnt_reviews = driver.find_element(
            By.CSS_SELECTOR, 
            'a.crema-product-reviews-count.crema-applied')

        title_list.append(title.text.strip())
        mean_rating_list.append(mean_rating.text.strip())
        enabled_size_list.append(enabled_size)
        foot_width_list.append(foot_width.text.strip())
        cnt_reviews_list.append(cnt_reviews.text.strip()[1:-1]+'개')

        driver.back()
        time.sleep(2)

except Exception as e :   # 에러 발생시 발생된 에러 출력
    print("에러명 :", e)
```

```python
newbal_dict = {"상품명":title_list, "평점":mean_rating_list, "리뷰개수":cnt_reviews_list,
               "구매가능 사이즈":enabled_size_list, "발볼넓이":foot_width_list}
newbal_df = pd.DataFrame(newbal_dict, index=range(1, len(title_list)+1))
newbal_df.index.name = "순위"
newbal_df
```

```python
# 컬럼 내부 너비 최대치로 옵션 변경
pd.set_option('display.max_colwidth', None)
newbal_df
```

### 스크롤 내려서 더 많은 영상 정보를 수집하기
- 원하는 개수만큼 수집해보자!

```python
# 영상 제목과 조회수 HTML 요소가 담길 리스트
title_list_html, view_cnt_html = [], []

# 수집 영상의 HTML 요소 개수가 50개 이상이 될 때까지 스크롤 내리기 반복
while len(title_list_html) < 50:
    body = driver.find_element(By.TAG_NAME, 'body')
    body.send_keys(Keys.END)
    time.sleep(3) # 영상이기 때문에 시간을 좀 더 넉넉히 두기

    title_list_html = driver.find_elements(By.CSS_SELECTOR, '#video-title > yt-formatted-string.style-scope.ytd-video-renderer')
    view_cnt_html = driver.find_elements(By.CSS_SELECTOR, '#metadata-line > span:nth-child(3)')

print('수집된 제목 개수:', len(title_list_html))
print('수집된 조회수 개수:', len(view_cnt_html))

# 우리가 원하는 컨텐츠(텍스트)만 빼서 리스트에 넣기
title_list, view_list = [], []

for i in tqdm(range(len(title_list_html))):
    title_list.append(title_list_html[i].text)
    view_list.append(view_cnt_html[i].text)

print('최종 영상 제목:', title_list)
print('최종 조회수:', view_list)
```

### **정확히 50개를 지정하여 추출해보자**
- 스크롤 횟수를 지정하는 것만으로는 정확한 데이터의 개수만큼 수집하는 것이 힘듦
- 스크롤은 약간 오버해서 내리더라도, 수집된 데이터 중에서 원하는 개수만큼만 가져오면 됨

```python
# 영상 제목과 조회수 HTML 요소가 담길 리스트
title_list_html, view_cnt_html = [], []

target_count = 50 # 우리가 수집을 원하는 데이터 개수

# 수집 영상의 HTML 요소 개수가 50개 이상이 될 때까지 스크롤 내리기 반복
while len(title_list_html) < target_count :
    body = driver.find_element(By.TAG_NAME, 'body')
    body.send_keys(Keys.END)
    time.sleep(3) # 영상이기 때문에 시간을 좀 더 넉넉히 두기

    title_list_html = driver.find_elements(By.CSS_SELECTOR, '#video-title > yt-formatted-string.style-scope.ytd-video-renderer')
    view_cnt_html = driver.find_elements(By.CSS_SELECTOR, '#metadata-line > span:nth-child(3)')

print('수집된 제목 개수:', len(title_list_html))
print('수집된 조회수 개수:', len(view_cnt_html))

# 우리가 원하는 컨텐츠(텍스트)만 빼서 리스트에 넣기
title_list, view_list = [], []

for i in tqdm(range(target_count)):
    title_list.append(title_list_html[i].text)
    view_list.append(view_cnt_html[i].text)

print('최종 영상 제목 수:', len(title_list))
print('최종 조회수 수 :', len(view_list))
```

### DF로 만들기

```python
youtube_df = pd.DataFrame({'제목': title_list, '조회수': view_list},
                         index = range(1, len(title_list)+1))
youtube_df.index.name = 'No'
youtube_df
```

### '시청 중'이라는 문자가 있는 행을 지우기
- 특정 문자가 있는 행을 지우려면 여러번 작업을 해야 하는데, 반대로 생각해서 접근
- 특정 문자가 없는 행만 남긴다! -> 한 줄로 작업 가능

```python
youtube_df = youtube_df[~youtube_df['조회수'].str.contains('시청 중')]
youtube_df
```
