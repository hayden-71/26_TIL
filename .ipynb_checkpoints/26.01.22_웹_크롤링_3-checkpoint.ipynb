# 네이버 블로그 제목 및 본문 내용 수집

```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
U_A = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36"}
```

```python
keyword = '현대자동차'
url = f"https://search.naver.com/search.naver?ssc=tab.blog.all&sm=tab_jum&query={keyword}"
```

```python
# 1. 통신 설정
# 2. 블로그 목록 정보 수집(타이틀, 블로거명, 게시일)
# 3. 해당 정보들로 데이터프레임 생성
# 4. 모든 코드 함수로 통합시키기
# (매개변수는 검색어가 입력되도록 설정하고 return은 데이터프레임으로!!)

res = req.get(url, headers=U_A)

soup = bs(res.text, 'lxml')
soup
```

```python
title = soup.select('span.sds-comps-text.sds-comps-text-ellipsis.sds-comps-text-ellipsis-1.sds-comps-text-type-headline1.sds-comps-text-weight-sm')
title[2].text
len(title)
```

```python
name = soup.select('a.fender-ui_228e3bd1.fender-ui_475445f0')
name[0].text
len(name)
```

```python
date = soup.select('span.sds-comps-text.sds-comps-text-type-body2.sds-comps-text-weight-sm.sds-comps-profile-info-subtext')
date[2].text
len(date)
```

```python
title_list, name_list, date_list = [], [], []

for i in range(len(title)):
    title_list.append(title[i].text.strip())
    name_list.append(name[i].text.strip())
    date_list.append(date[i].text.strip())

print(title[1].text + name[1].text + date[1].text)
```

```python
blog_dic = {'타이틀':title_list, '블로거명':name_list, '게시일':date_list}

blog_df = pd.DataFrame(blog_dic, index=range(1,len(title_list)+1))
blog_df.index.name = 'No'
blog_df
```

```python
def search(keyword:str) -> pd.DataFrame:
    title_list, name_list, date_list = [], [], []

    U_A = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36"}
    url = f"https://search.naver.com/search.naver?ssc=tab.blog.all&sm=tab_jum&query={keyword}"
    res = req.get(url, headers=U_A)
    soup = bs(res.text, 'lxml')

    title = soup.select('span.sds-comps-text.sds-comps-text-ellipsis.sds-comps-text-ellipsis-1.sds-comps-text-type-headline1.sds-comps-text-weight-sm')
    name = soup.select('a.fender-ui_228e3bd1.fender-ui_475445f0')
    date = soup.select('span.sds-comps-text.sds-comps-text-type-body2.sds-comps-text-weight-sm.sds-comps-profile-info-subtext')

    for i in range(len(title)):
        title_list.append(title[i].text.strip())
        name_list.append(name[i].text.strip())
        date_list.append(date[i].text.strip())

        blog_dic = {'타이틀':title_list, '블로거명':name_list, '게시일':date_list}

    blog_df = pd.DataFrame(blog_dic, index=range(1,len(title_list)+1))
    blog_df.index.name = 'No'
    return blog_df


search('전자상가 타건')
```
