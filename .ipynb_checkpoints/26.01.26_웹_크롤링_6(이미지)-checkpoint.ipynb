# 네이버에서 이미지 데이터 수집하기

```python
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
from tqdm import tqdm

from bs4 import BeautifulSoup as bs
# 파일 및 폴더 관리 라이브러리 
import os
# urlretrieve: url 경로를 통해 데이터를 파일로 다운시켜주는 클래스 
from urllib.request import urlretrieve
```

## 수집된 이미지가 저장될 폴더 설정
- 폴더가 없으면 생성되도록 설정

```python
# isdir: 폴더가 있는지 확인
if not os.path.isdir('이미지'):
    os.mkdir('이미지') # mkdir: 폴더 생성
    print('해당 경로에 폴더를 생성합니다~')
else:
    print('해당 경로에 폴더가 존재합니다~')


# 자주 사용하는 리눅스 명령
 # cd: 경로 변경
 # ls: 파일명 확인 (파이썬에서는 isdir 활용)
 # rm: 지우기
 # mkdir: 폴더 생성 
```

## 네이버 검색창에 원하는 검색어로 검색 후, 이미지 탭으로 들어가서 url 복사하기

```python
driver = wb.Chrome()
driver.maximize_window()
time.sleep(1)

driver.get('https://search.naver.com/search.naver?ssc=tab.image.all&where=image&sm=tab_jum&query=%EB%AC%B4%ED%95%9C%EB%8F%84%EC%A0%84+%EC%A7%A4')
```

## 이미지 요소 추출
## 1) Beautiful Soup 활용

```python
# page_source: 페이지의 HTML 소스 정보를 가져오기
soup = bs(driver.page_source, 'lxml')
soup
```

```python
# 이미지들의 HTML 요소에 접근
imgs_bs = soup.select('img._fe_image_tab_content_thumbnail_image')
imgs_bs[0].text # 이미지 태그 내에 따로 텍스트는 없쥬
```

```python
len(imgs_bs)
```

```python
imgs_bs

# 이미지는 실제 이미지가 들어있는 서버 주소인 src 속성 값으로 접근해야 함
```

```python
# BS로 속성에 접근하기 위해서는 [] 사용!!
imgs_bs[0]['src']
```

## 2) Selemium 단독 활용 

```python
imgs_sel = driver.find_elements(
    By.CSS_SELECTOR, 'img._fe_image_tab_content_thumbnail_image')

len(imgs_sel)
```

```python
imgs_sel[0].get_attribute('src')
```

```python
# 이미지들의 주소만 리스트에 넣기
imgs_src_list = [i.get_attribute('src') for i in imgs_sel]
imgs_src_list

# imgs_src_list = []
# for i in imgs_sel:
#     imgs_src_list.append(i.get_attribute('src'))
```

- 밑쪽에 gif;로 표시된 이미지들은 HTML 요소에는 접근했지만, 아직 이미지의 서버 주소는 로드되지 않은 상태

```python
# 조건문으로 gif 외에 https로 시작되는 주소만 걸러내서 작업하기
                                    # if i.startswith('https') 혹은 if 'https' in i
img_src_list_new = [i for i in imgs_src_list if i[:5]=='https']
img_src_list_new

# img_src_list_new = []
# for i in imgs_src_list:
#     if i.startswith('https'):
#         img_src_list_new.append(i.get_attribute('src'))
```

```python
len(imgs_src_list), len(img_src_list_new)
```

## 각 이미지들을 파일로 변환 후 생성한 폴더에 담기

```python
for idx, img in tqdm(enumerate(img_src_list_new)):
    # urlretrieve(이미지 url, 다운받을 경로 및 확장자)
    urlretrieve(img, '이미지/' + str(idx+1) + '.jpg')
```

## 전체 코드를 한 셀에 작성하기
- 스크롤 기능 사용하여 3번 스크롤 내린 후, 이미지 수집하게 코드 작성

```python
driver = wb.Chrome()
driver.maximize_window()
time.sleep(2)

driver.get('https://search.naver.com/search.naver?ssc=tab.image.all&where=image&sm=tab_jum&query=%EB%AC%B4%ED%95%9C%EB%8F%84%EC%A0%84+%EC%A7%A4')


try:
    # 시간 텀 주어 스크롤 내리기
    for _ in range(3):
        body = driver.find_element(By.TAG_NAME, 'body')
        body.send_keys(Keys.END)
        time.sleep(2)

    # 이미지 요소에 접근
    imgs_sel = driver.find_elements(
    By.CSS_SELECTOR, 'img._fe_image_tab_content_thumbnail_image')

    # 수집된 요소의 src 속성만 추출하여 리스트에 저장
    imgs_sel_list = [i.get_attribute('src') for i in imgs_sel]
    new_imgs_list = [i for i in imgs_sel_list if i[:5]=='https']

    # 수집된 링크로 이미지 파일 다운로드
    for idx, img in tqdm(enumerate(new_imgs_list)):
        urlretrieve(img, '이미지/' + str(idx+100) + '.jpg')
    time.sleep(0.5) # 이미지 다운받는 속도 감안하여 조금씩 텀 두기

    print('크롤링 완!!')
    driver.quit()

except Exception as e:
    print('에러명:', e)
```

```python
driver = wb.Chrome()
driver.get('https://naver.com')
keyword = input('검색 키워드 입력: ')

driver.get(f'https://search.naver.com/search.naver?ssc=tab.image.all&where=image&sm=tab_jum&query={keyword}')
```

## **이미지 데이터가 제대로 수집되지 않을수 있는 이유들**
1. 웹 페이지 자체의 실시간 변동성이 심한 경우
   - 빈번한 업데이트, 업로드, 삭제 등으로 수집하려는 대상이 동작 중 변경될 수 있음
2. 웹 페이지별로 구조가 상이한 경우
   - 반응형 웹과 같이 상황에 따라 구조가 변하는 경우, 요소가 있다가 없어지는 현상이 발생할 수 있음
3. 크롤링 시 충분한 시간 텀을 주지 않은 경우
   - 웹 페이지 이미지가 다 로드되기도 전에 코드만 동작시키면 제대로 인식되지 않을 수 있음
4. 수집하려는 데이터 자체가 부족한 경우
   - 희귀한 검색어의 경우 업로드 된 이미지 자체가 적어서 수집이 생각보다 덜 될 수 있음
