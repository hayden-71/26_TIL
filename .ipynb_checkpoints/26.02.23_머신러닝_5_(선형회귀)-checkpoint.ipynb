# 선형회귀 모델 사용 및 w, b 값 확인

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression # 선형회귀 모델 클래스
```

```python
x = range(5)
y = [2,4,6,8,10]
```

```python
plt.figure(figsize=(5,3))
plt.scatter(x, y)
plt.grid()
plt.show()
```

```python
# 모델 학습을 위한 데이터 형식 변경 (DF화)
X_train = pd.DataFrame(x)
y_train = pd.DataFrame(y)
```

```python
# 선형회귀모델 객체 생성
lr_model = LinearRegression()
```

```python
lr_model.fit(X_train, y_train)
```

```python
lr_model.predict([[2.5]])  # DF형태로 학습시켰으니, 2차원으로 넣어주기
```

```python
print('w값 :', lr_model.coef_)  # w값 (가중치, 회귀계수 coefficient)
print('b값 :', lr_model.intercept_)  # b값 (y절편 intercept)
```

```python
import seaborn as sns
from sklearn.model_selection import train_test_split
```

### **데이터 수집**

```python
data = pd.read_csv('data/boston_house.csv')
data
```

```python
data.info()
```

### **데이터 전처리**
- 결측치, 이상치는 없음
- 아래에서 특성확장 진행 예정

### **탐색적 데이터 분석**

```python
# 상관계수 출력
data.corr()
```

```python
plt.figure(figsize=(15,15))

sns.heatmap(
    data=data.corr(),
    annot=True,
    fmt='.2f',
    cmap='Blues'
)
```

- 주택가격(MEDV)과 상관관계가 높은 특성은 주택당 방 개수(RM), 소득하위 계층비율(LSTAT)

### **모델링 (데이터 가공, 모델 선택, 학습, 평가)**

```python
# 문제, 정답 및 학습, 평가 데이터로 분리하고 선형회귀 모델로 학습시켜보세요
y = data['MEDV']
X = data.drop('MEDV', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=5)

# 정답이 실수인 데이터는 stratify=y 를 사용할 수 없음 (class가 없기 때문에)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# MSE(평균제곱오차) 평가 지표 클래스
from sklearn.metrics import mean_squared_error
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
```

```python
# feature_names_in
# 학습된 선형회귀 모델에서 데이터 특성 명칭 출력
# y = w1x1 + w2x2 + ... + w13x13 + b
lr_model.feature_names_in_
```

```python
# 특성 개수(13개)만큼 w값(가중치, 회귀계수) 출력
lr_model.coef_
```

```python
# y 절편 (b값)
lr_model.intercept_
```

- 위 w값들 중 절대값이 비교적 큰 'RM' 컬럼은 상관계수도 컸지만, 'LSTAT'컬럼은 w값이 작았음
- 정답과의 상관계수가 높을수록 해당 특성의 w값이 커지는 경향이 있지만, 다른 특성과의 조합에 따라 w값들이 변동되기 때문에
- 상관계수가 높다고 해서 해당 특성의 w가 반드시 높다고 볼 수는 없음
- 즉, `상관계수는 두 변수간의 상관성 정도`이고, `w는 입력되는 모든 특성을 고려한 중요도`

### MSE, RMSE 평가지표 확인

```python
pred_train = lr_model.predict(X_train)
pred_test = lr_model.predict(X_test)
```

```python
pred_train
```

```python
# MSE, RMSE 수치 확인
train_mse = mean_squared_error(y_train, pred_train) # (실제정답, 예측값)
test_mse = mean_squared_error(y_test, pred_test)

print('train 데이터 MSE:', train_mse)
print('test 데이터 MSE:', test_mse)
print('train, test의 MSE 차이:', abs(train_mse-test_mse))

print('='*40)

print('train 데이터 RMSE:', train_mse**0.5)      # np.sqrt 사용해도 됨
print('test 데이터 RMSE:', test_mse**0.5)
print('train, test의 RMSE 차이:', abs(train_mse**0.5 - test_mse**0.5))
```

```python
print('train 데이터 주택 가격 평균:', y_train.mean())
print('test 데이터 주택 가격 평균:', y_test.mean())
```

- test 데이터의 주택 가격 평균은 22,000달러인데 전체 오차인 RMSE가 5,500달러면 전체의 25% 수준
- 상당히 큰 오차

### R2 스코어 확인

```python
# 회귀 모델의 score함수는 R2_socre가 출력됨 (분류 모델은 accuracy)
print('train R2 socre:', lr_model.score(X_train, y_train))
print('test R2 socre:', lr_model.score(X_test, y_test))

# R2 score를 회귀결정계수 라고 부르기도 함
```

### 평가 함수 만들어서 활용 

```python
# MSE, RMSE, R2_Score 평가 함수 선언
# (모델명, 학습용 문제 데이터, 평가용 문제 데이터)
def evaluate_model(model, X_train, X_test) : 
    pred_train = model.predict(X_train)
    pred_test = model.predict(X_test)

    train_mse = mean_squared_error(y_train, pred_train)
    test_mse = mean_squared_error(y_test, pred_test)
    
    print('train 데이터 MSE:', train_mse)
    print('test 데이터 MSE:', test_mse)
    print('train, test의 MSE 차이:', abs(train_mse-test_mse))
    
    print('='*40)
    
    print('train 데이터 RMSE:', train_mse**0.5)      # np.sqrt 사용해도 됨
    print('test 데이터 RMSE:', test_mse**0.5)
    print('train, test의 RMSE 차이:', abs(train_mse**0.5 - test_mse**0.5))
    
    print('='*40)

    print('train R2 socre:', model.score(X_train, y_train))
    print('test R2 socre:', model.score(X_test, y_test))
```

```python
evaluate_model(lr_model, X_train, X_test)
```

## **선형회귀모델의 데이터 특성 확장**
- 선형회귀모델은 데이터의 분포가 선형 형태가 아니라면 예측정확도가 낮아짐. 모델 튜닝을 위한 하이퍼파라미터가 없음
- 즉, 데이터 분포 자체에 영향을 많이 받게 됨
- 다중선형회귀(특성 2개 이상)에서 일반적으로 특성의 개수가 적으면(데이터 복잡도가 낮으면) 하이퍼파라미터 튜닝으로 성능을 높일 수 있는 다른 모델들에 비해, 선형회귀모델은 한계가 있음
- 그래서, 데이터단에서 **새로운 특성들을 추가하여 특성의 개수를 늘려 데이터의 복잡도를 증가**시키는 기법 사용

```python
X_train.columns
```

```python
# DF를 만들 데이터를 넣기 위한 빈 딕셔너리 생성
extended_train_col = {}

# 각 컬럼들을 서로서로 한 번씩 곱해서 만든 컬럼을 딕셔너리에 추가 
for col1 in X_train.columns :
    for col2 in X_train.columns :
        extended_train_col[col1 + 'x' + col2] = X_train[col1] * X_train[col2]
extended_train_col
```

```python
# 딕셔너리 df화 시키기
extended_X_train_df = pd.DataFrame(extended_train_col)

# 기존 컬럼에 확장된 컬럼 붙여주기
extended_X_train = pd.concat([X_train, extended_X_train_df], axis=1)
extended_X_train
```

```python
# test 데이터도 똑같이 진행
extended_test_col = {}

# 각 컬럼들을 서로서로 한 번씩 곱해서 만든 컬럼을 딕셔너리에 추가 
for col1 in X_test.columns :
    for col2 in X_test.columns :
        extended_test_col[col1 + 'x' + col2] = X_test[col1] * X_test[col2]
extended_test_col
```

```python
extended_X_test_df = pd.DataFrame(extended_test_col)

# 기존 컬럼에 확장된 컬럼 붙여주기
extended_X_test = pd.concat([X_test, extended_X_test_df], axis=1)
extended_X_test
```

```python
# ========== 이런 방법도 있어 ===============
# PolynomialFeatures(다항 특성): 데이터 재료 뻥튀기,,

from sklearn.preprocessing import PolynomialFeatures

# 1. 도구 준비 (degree=2는 2차항까지 만들겠다는 뜻)
poly = PolynomialFeatures(degree=2, include_bias=False)

# 2. 데이터 변신! (fit_transform)
# 예: [2, 3] 이라는 데이터가 들어간다면?
X_poly = poly.fit_transform([[2, 3]])

print(X_poly)
# 결과: [[2.0, 3.0, 4.0, 6.0, 9.0]]
# 해석: [a, b, a^2, a*b, b^2] 순서대로 생겨난 거예요!


# ==============================
# [준비물]
alphas = [0.01, 0.1, 1, 10]
best_rmse = float('inf') # 아주 큰 숫자로 시작
best_alpha = None

# --- [여기서부터가 반복 노가다 시작] ---
for a in alphas:
    # 1. 데이터 특성 확장 (재료 뻥튀기)
    poly = PolynomialFeatures(degree=2)
    X_train_poly = poly.fit_transform(X_train)
    
    # 2. 데이터 스케일링 (크기 맞추기)
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train_poly)
    
    # 3. 모델 만들고 교차 검증 (연습 시험 5번 보기)
    model = ElasticNet(alpha=a)
    # 교차 검증 함수가 내부적으로 데이터를 5개로 쪼개서 점수를 내줌
    scores = cross_val_score(model, X_train_scaled, y_train, cv=5, scoring='neg_root_mean_squared_error')
    avg_rmse = np.mean(-scores) 
    
    print(f"알파 {a}일 때 모의고사 평균 점수: {avg_rmse}")

    # 4. 가장 좋은 알파값 기억하기
    if avg_rmse < best_rmse:
        best_rmse = avg_rmse
        best_alpha = a

# --- [반복 끝! 이제 최고의 알파를 찾았음] ---

print(f"최종 결정된 최고의 알파: {best_alpha}")

# 5. [재학습 - Re-fit] 찾은 최고의 알파로 '전체 데이터' 학습하기
final_poly = PolynomialFeatures(degree=2)
X_final_train = final_poly.fit_transform(X_train) # 전체 훈련 데이터 뻥튀기

final_scaler = StandardScaler()
X_final_scaled = final_scaler.fit_transform(X_final_train) # 전체 데이터 스케일링

final_model = ElasticNet(alpha=best_alpha)
final_model.fit(X_final_scaled, y_train) # 드디어 전체 데이터로 '최종 합체' 학습!

# 6. [실전] 드디어 딱 한 번, 테스트 데이터로 시험 보기
X_test_poly = final_poly.transform(X_test)
X_test_scaled = final_scaler.transform(X_test_poly)
final_pred = final_model.predict(X_test_scaled)
```

- 특성확장으로 성능이 크게 올라간 것 확인

## **라쏘(Lasso) 모델 적용**
- 선형회귀에 L1 규제가 추가된 모델
- L1 규제는 w에 같은 값 (상수)을 빼주는 방식

```python
from sklearn.linear_model import Lasso
```

```python
lasso_model = Lasso(
    alpha=10,        # 규제의 강도
    max_iter=10000,   # 최적 w를 찾기 위한 학습 반복 횟수 
)
```

```python
lasso_model.fit(extended_X_train, y_train)
```

```python
# 라쏘 모델은 중요도가 낮은 특성의 w값을, 업데이트를 통해 0으로 만들어주기 때문에
# 0이 아닌 w값의 개수를 파악하면, 정답 예측에 몇 개의 특성을 사용했는지 알 수 있음
print(f'정답 예측에 사용한 특성 수: {sum(lasso_model.coef_ != 0)}')
```

```python
evaluate_model(lasso_model, extended_X_train, extended_X_test)
```

- train의 R2_Score는 조금 낮아졌지만, 선형회귀 모델에 비해 과대적합 꽤 해소 

## **릿지(Ridge) 모델 적용**
- 선형회귀에 L2 규제가 추가된 모델
- L2 규제는 모든특성의 w값을 같은 비율로 줄여줌


```python
from sklearn.linear_model import Ridge

ridge = Ridge(
    alpha=10,
    max_iter=10000
)
```

```python
ridge.fit(extended_X_train, y_train)
```

```python
evaluate_model(ridge, extended_X_train, extended_X_test)
```

- alpha값이 10일 때는 라쏘모델이 릿지에 비해 과대적합해소에 더 유리했음 

## **엘라스틱넷(ElasticNet) 모델 적용**
- L1과 L2 규제를 혼합해서 사용

```python
from sklearn.linear_model import ElasticNet

ela = ElasticNet(
    alpha=10,
    max_iter=10000,
    l1_ratio=0.5    # L1 규제의 비율 (1에서 뺀 나머지는 L2)
)

ela.fit(extended_X_train, y_train)
```

```python
evaluate_model(ela, extended_X_train, extended_X_test)
```

- 현재의 하이퍼파라미터 설정으로는 엘라스틱넷이 과대적합에 제일 유리
- 라쏘모델에 비해 R2_Score도 미세하게 높은 편

#### 경사하강법
- "미분값이 0이 되는 지점을 찾는 반복 알고리즘"
- loss = (정답 - 예측)^2
    - loss를 최소로 만드는 w를 찾고 싶음.
- w = w - learning_rate * 미분값
    - 지금 위치에서 기울기(미분값)를 본다
    - 내려가는 방향으로 한 걸음 이동한다
    - 반복하면서 바닥(최적 w)으로 내려간다
    - 미분값 = 0 → 기울기 없음 → 최소 지점

## 하이퍼파라미터 튜닝
- alpha 값을 바꿔가며 RMSE의 변화를 확인해보자!

```python
# Lasso 모델
alpha_list = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
lasso_train_list = []
lasso_test_list = []

for a in alpha_list : 
    lasso = Lasso(alpha = a, max_iter=10000)      # 모델 객체 생성
    lasso.fit(extended_X_train, y_train)    # 모델 학습

    lasso_train_pred = lasso.predict(extended_X_train)      # 예측
    lasso_train_RMSE = mean_squared_error(y_train, lasso_train_pred) **0.5   # RMSE 전체에 ** 0.5
    lasso_train_list.append(lasso_train_RMSE)    # 리스트에 RMSE 결과 담기

    lasso_test_pred = lasso.predict(extended_X_test)
    lasso_test_RMSE = mean_squared_error(y_test, lasso_test_pred) **0.5
    lasso_test_list.append(lasso_test_RMSE)
```

```python
plt.figure(figsize=(10,4))
plt.plot(lasso_train_list, label='train')
plt.plot(lasso_test_list, label='test')
plt.title('Lasso : alpha - RMSE chart')
plt.xlabel('alpha')
plt.ylabel('RMSE')
plt.xticks(np.arange(7), alpha_list)   # x축 표시값 설정
plt.ylim(2, 6.5)
plt.legend()
plt.grid()
plt.show
```

```python
# Ridge 모델
alpha_list = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
ridge_train_list = []
ridge_test_list = []

for a in alpha_list : 
    ridge = Ridge(alpha = a, max_iter=10000)      # 모델 객체 생성
    ridge.fit(extended_X_train, y_train)    # 모델 학습

    ridge_train_pred = ridge.predict(extended_X_train)      # 예측
    ridge_train_RMSE = mean_squared_error(y_train, ridge_train_pred) **0.5   # RMSE 전체에 ** 0.5
    ridge_train_list.append(ridge_train_RMSE)    # 리스트에 RMSE 결과 담기

    ridge_test_pred = ridge.predict(extended_X_test)
    ridge_test_RMSE = mean_squared_error(y_test, ridge_test_pred) **0.5
    ridge_test_list.append(ridge_test_RMSE)
```

```python
plt.figure(figsize=(10,4))
plt.plot(ridge_train_list, label='train')
plt.plot(ridge_test_list, label='test')
plt.title('ridge : alpha - RMSE chart')
plt.xlabel('alpha')
plt.ylabel('RMSE')
plt.xticks(np.arange(7), alpha_list)   # x축 표시값 설정
plt.ylim(2, 6.5)
plt.legend()
plt.grid()
plt.show
```

### 결론
- 현재 데이터에서는 alpha가 1이나 10일 때,  Ridge에 비해 lasso 모델의 train, test의 RMSE 차이가 더 적음
- 과대적합에 유리하다고 볼 수 있음

## 앙상블 회귀 모델 적용

```python
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

### **1) 랜덤 포레스트**

```python
rfr = RandomForestRegressor(
    n_estimators=1000,
    max_depth=7,
    n_jobs=-1,      # 모든 CPU 코어 사용 (병렬학습 가능)
    random_state=11
)

rfr.fit(extended_X_train, y_train)
```

```python
evaluate_model(rfr, extended_X_train, extended_X_test)
```

### **2) XG부스팅**
- 경사하강법 적용

```python
xgb = XGBRegressor(
    n_estimators=1000,
    learning_rate=0.01,    # 학습률 (경사하강법 업데이트 간격)
    max_depth=7,
    random_state=11,
    n_jobs=-1
)

xgb.fit(extended_X_train, y_train)
```

```python
evaluate_model(xgb, extended_X_train, extended_X_test)
```
